# Example illustrating the use of the dual beta-process, Gaussian process (DBP-GP) algorithm

```python
%matplotlib inline

import numpy as np
from numpy import linspace
from matplotlib import pyplot as plt
from neu_ml.code.data.sample_polynomial import sample_polynomial
from neu_ml.code.utils.compute_gram_mat import compute_gram_mat
from neu_ml.code.algorithms.regression.dual_bp_gp_regression \
    import dual_bp_gp_regression
from neu_ml.code.visualization.plot_latent_trajectories \
    import plot_latent_trajectories
    
np.set_printoptions(precision = 3, suppress = True, linewidth=200)
```

Begin by defining some parameter values

```python
N = 60 # Number of instances
D = 4 # Number of features
K = 20 # Truncation level of the beta-process
var = .2 # The variance of each feature
```

Generate some data to pass to the algorithm

```python
line1 = sample_polynomial(n_points = N/2, cos = np.array([2., 0.]), \
                    domain_range = np.array([0., 10.]), var = var)

line2 = sample_polynomial(n_points = N/2, cos = np.array([-2., 20.]), \
                    domain_range = np.array([0., 10.]), var = var)

line3 = sample_polynomial(n_points = N/2, cos = np.array([2., 10.]), \
                    domain_locs = line1[:, 0], var = var)

line4 = sample_polynomial(n_points = N/2, cos = np.array([-2., 10.]), \
                    domain_locs = line2[:, 0], var = var)

x = np.vstack((line1, line2))[:, 0]
y = np.zeros([N, 4])

line5 = sample_polynomial(n_points = N, cos = np.array([0., 0.]), \
                    domain_locs = x, var = var)

tmp = linspace(0, N-1, N)%3 == 0
locs6 = linspace(0, N-1, N).astype(int)[tmp]
line6 = sample_polynomial(n_points = N/3, cos = np.array([-1., 20.]), \
                        domain_locs = x[locs6], var = var)

tmp = linspace(0, N-1, N)%3 == 1
locs7 = linspace(0, N-1, N).astype(int)[tmp]
line7 = sample_polynomial(n_points = N/3, cos = np.array([-4., 20.]), \
                        domain_locs = x[locs7], var = var)

tmp = linspace(0, N-1, N)%3 == 2
locs8 = linspace(0, N-1, N).astype(int)[tmp]
line8 = sample_polynomial(n_points = N/3, cos = np.array([-8., 20.]), \
                        domain_locs = x[locs8], var = var)

y[0:N/2, 0] = line1[:, 1]
y[N/2:N, 0] = line2[:, 1]
y[0:N/2, 1] = line3[:, 1]
y[N/2:N, 1] = line4[:, 1]
y[:, 2] = line5[:, 1]    
y[locs6, 3] = line6[:, 1]
y[locs7, 3] = line7[:, 1]
y[locs8, 3] = line8[:, 1]    
```

Now let's plot the data

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(nrows=1, ncols=4, sharex=False, sharey=False)
fig.tight_layout()
fig.set_figwidth(20)

ax1.scatter(x, y[:, 0])
ax2.scatter(x, y[:, 1])
ax3.scatter(x, y[:, 2])
ax4.scatter(x, y[:, 3])
```

Now set up the parameters that control the beta processes for both the instance-latent cluster association and the feature-latent cluster association.

```python
alpha_instances = 3.5
gamma_instances = 1.5
ids_instacnes = np.linspace(1, N, N) - 1
instance_bp_param_groups = [{'alpha': alpha_instances, \
                            'gamma': gamma_instances, \
                            'ids': ids_instacnes}]

alpha_features = 1.5
gamma_features = 1.5
ids_features = np.linspace(1, D, D) - 1
feature_bp_param_groups = [{'alpha': alpha_features, \
                            'gamma': gamma_features, \
                            'ids': ids_features}]    
```

Set up the linear kernel to use for the Gram matrices needed by the Gaussian processes. We need one for each of the four observed features. For this simple example, we use the same Gram matrix for each feature.

```python
kernel_name   = 'linear'
kernel_params = np.array([2.0, 1.5, 0.0])
kernel = [kernel_name, kernel_params]

Ks = np.zeros([D, N, N])
Ks[0, :, :] = Ks[1, :, :] = Ks[2, :, :] = Ks[3, :, :] = \
    compute_gram_mat(x, kernel)
```

We can set the signal means ('sig_means') and signal precision ('sig_betas') to be used by the Gaussian process component of the likelihood term. We can also set the noise means ('noise_means') and noise precision values ('noise_betas') to be used by the noise model component of the likelihood.

```python
sig_means = np.zeros([N, D])
sig_means[:, 0] = 4.5*x
sig_means[:, 1] = 10.
sig_means[:, 2] = 0.
sig_means[:, 3] = -2.5*x + 20.

sig_betas = (1./var)*np.ones(D)

noise_means = np.ones([N, D])
noise_means[:, 0] = np.mean(y[:, 0])
noise_means[:, 1] = np.mean(y[:, 1])
noise_means[:, 2] = np.mean(y[:, 2])
noise_means[:, 3] = np.mean(y[:, 3])    

noise_betas = sig_betas
```

Finally, run the algorithm. By default the algorithm is run with random initializations for both 'phi_instances' (matrix of Bernoulli parameters indicating the association betwen instances and latent clusters) and 'phi_featuers' (matrix of Bernoulli parameters indicating the association between features and latent clusters), although these can be passed to the function if the user has a general idea about what they should be. Also, setting the 'verbose_lb' option to True allows us to monitor the increase in the variation lower bound as the algorithm executes, but this also cause the algorithm to run longer. The user also has the opportunity to sepecify the number of repeats, the number of times the algorithm is re-run in order to find the best lower bound value. In practice it may a number of runs to find a satisfactory solution.

```python
phi_instances, phi_features, lower_bound = \
    dual_bp_gp_regression(data = y, Ks = Ks,
        sig_means = sig_means, sig_betas = sig_betas,
        noise_means = noise_means, noise_betas = noise_betas,            
        instance_bp_param_groups = instance_bp_param_groups,
        feature_bp_param_groups = feature_bp_param_groups,
        iters = 10, K = K, prob_thresh = 1e-10, verbose_lb=True)
```
